```python
df['total_score'] = df['math_score'] + df['writing_score'] + df['reading_score']
df['avg_score'] = df['total_score']/3
df.head()
```

```python
reading_full = df[df['reading_score']==100]['avg_score'].count()
writing_full = df[df['writing_score']==100]['avg_score'].count()
math_full = df[df['math_score']==100]['avg_score'].count()
print("reading full marks total student :",reading_full)
print("writing full marks total student :",writing_full)
print("math full marks total student    :",math_full)
```

```python
reading_u30 = df[df['reading_score']<=30]['avg_score'].count()
writing_u30 = df[df['writing_score']<=30]['avg_score'].count()
math_u30 = df[df['math_score']<=30]['avg_score'].count()
print("reading under 20% marks total student :",reading_u30)
print("writing under 20% marks total student :",writing_u30)
print("math under 20% marks total student    :",math_u30)
```

from above data analysis we can say that most student <br>
have performed good in reading whereas worst performance in math

```python
pd.options.display.max_seq_items = 2000

gender_group = df.groupby(['gender']).sum()
print(gender_group) #['math_score', 'reading_score',  'writing_score',  'total_score', 'avg_score']
```

female have performed better than male students <br>
male have score a bit better in  maths, in other subjects female dominance is there

```python
plt.figure(figsize=(15,6))

X = ['total avg', 'maths avg']

female_scores = [gender_group['avg_score'][0], gender_group['math_score'][0]]
male_scores = [gender_group['avg_score'][1], gender_group['math_score'][1]]

print(male_scores, female_scores)

X_axis = np.arange(len(X))
plt.bar(X_axis-0.2, male_scores, 0.39, label='Male') #having width=0.4 will attach both bars
plt.bar(X_axis+0.2, female_scores, 0.39, label='Female')

plt.xticks(X_axis,X)
plt.ylabel("MArks scored")
plt.title("score comparison between male and female students")
plt.legend()
plt.show()
```

```python
Group_data2=df.groupby('race_ethnicity')

f,ax=plt.subplots(1,3,figsize=(20,8))

sns.barplot(x=Group_data2['math_score'].mean().index,y=Group_data2['math_score'].mean().values, palette='mako', ax=ax[0])
ax[0].set_title('Math score', size=20)
for container in ax[0].containers:
    ax[0].bar_label(container,color='black',size=15)


sns.barplot(x=Group_data2['reading_score'].mean().index,y=Group_data2['reading_score'].mean().values, palette='flare', ax=ax[1])
ax[1].set_title('Reading score', size=20)

for container in ax[1].containers:
    ax[1].bar_label(container,color='black',size=15)


sns.barplot(x=Group_data2['writing_score'].mean().index,y=Group_data2['writing_score'].mean().values, palette='dark:salmon_r', ax=ax[2])
ax[2].set_title('Writing score', size=20)
for container in ax[2].containers:
    ax[2].bar_label(container,color='black',size=15)


plt.show()
```

group E have performed better in all the subjects <br>
while group A,B have lower scores <br>
This shows that students from lower-socioeconomic class have performed good
